```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
data = pd.read_csv("IBM Attrition Data.csv")
data.head()
```

```python
data.describe()
```

```python
print(data.shape)
print(data.columns)
```

```python
style.use('ggplot')
plt.figure(figsize=(8,6))
plt.hist(data['Age'], bins=50, color='b')
plt.xlabel("Age")
plt.ylabel("Number of persons in range")
plt.title("Age Distribution of Employees")
plt.show()
```

```python
style.use('ggplot')
data_attr1 = data[data["Attrition"]=="Yes"] # attrition 1 dataset
data_attr2 = data[data["Attrition"]=="No"] 
plt.scatter(data_attr1["Age"], data_attr1["Attrition"])
plt.scatter(data_attr2["Age"], data_attr2["Attrition"])
plt.show()
# data points are almost dimstributed uniformly which might mean that this feature will be almost indecisive
# to determine attrition of an employee
```

```python
# number of employees having attrition as yes 
plt.figure(figsize = (40, 20))
labels = data_attr1['EducationField'].unique()
plt.subplot(1, 3, 1)
style.use("ggplot")
num = []
for label in labels:
    num.append(sum(data["EducationField"]==label))
plt.pie(num, labels=labels, autopct="%1.1f")
plt.title("Employees distribution")

plt.subplot(1, 3, 2)
style.use("ggplot")
num_attr1 = []
for label in labels:
    num_attr1.append(sum(data_attr1['EducationField']==label))
plt.pie(num_attr1, labels = labels, autopct="%1.1f")
plt.title("Employees having attrition YES")

plt.subplot(1, 3, 3)
style.use("ggplot")
num_attr2 = []
for label in labels:
    num_attr2.append(sum(data_attr2['EducationField']==label))
plt.pie(num_attr2, labels = labels, autopct="%1.1f")
plt.title("Employees having attrition NO")

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
style.use("ggplot")
plt.hist(data['MaritalStatus'], bins=5)

plt.subplot(1, 2, 2)
style.use("ggplot")
num = []
labels = data["MaritalStatus"].unique()
for label in labels:
    num.append(sum(data['MaritalStatus']==label))
plt.pie(num, labels=labels, autopct='%1.1f')
plt.suptitle("Marital Status of Employees")
plt.show()
```
